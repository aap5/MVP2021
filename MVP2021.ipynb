```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')  ##For easier viewing of visualizations
import numpy as np
```

```python
url = 'https://www.basketball-reference.com/awards/awards_2021.html'    #Where data is being pulled from
```

```python
df = pd.read_html(url, header = 1)   
df
```

```python
mvp2021 = df[0]
mvp2021["MVPvotes"] = mvp2021["Pts Won"] / mvp2021["Pts Max"] ##Percentage of total points awarded to an MVP canidate
mvp2021["Defense"] = mvp2021["BLK"].round(0) + mvp2021["STL"].round(0) #Combining defensive stats steals and blocks
mvp2021age = mvp2021.sort_values(by = "Age",ascending = True).reset_index() ##Sorting by age
mvp2021ptsWon = mvp2021.sort_values(by = "Pts Won",ascending = False).reset_index() ##Sorting by MVP votes
mvp2021games = mvp2021.sort_values(by = "G",ascending = True).reset_index() ##Sorting by MVP votes
mvp2021d = mvp2021.sort_values(by = "Defense",ascending = True).reset_index() ##Sorting by MVP votes
mvp2021games
```

```python
mvp2021age.plot.bar( y = "Pts Won", x = "Age", ylabel = "MVP Points Won", title ="MVP votes by age" )  ##The MVP votes by age
```

It appears that younger players receie more MVP votes
